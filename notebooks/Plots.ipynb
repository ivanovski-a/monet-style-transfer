## Our model training loss vs. epochs

```python
def plot_training_losses_gen(gen_losses, subtitle):
    epochs = list(range(1, len(gen_losses) + 1))
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, gen_losses, label='Generator Loss', color='blue')

    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.suptitle(f'Training Losses over {len(epochs)} Epochs - Generators', fontsize=14)
    plt.title(subtitle, fontsize=12)

    plt.xticks(range(1, len(epochs)+1, 2))

    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_training_losses_disc(disc_loss_M, disc_loss_P, subtitle):
    epochs = list(range(1, len(disc_loss_M) + 1))
    plt.figure(figsize=(12, 6))
    plt.plot(epochs, disc_loss_M, label='Discriminator Loss M', color='red')
    plt.plot(epochs, disc_loss_P, label='Discriminator Loss P', color='green')

    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.suptitle(f'Training Losses over {len(epochs)} Epochs - Discriminators', fontsize=14)
    plt.title(subtitle, fontsize=12)

    plt.xticks(range(1, len(epochs)+1, 2))

    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
